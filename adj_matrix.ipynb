```python
import geopandas as gpd
import matplotlib.pyplot as plt
japanshape = gpd.read_file('./gadm36_JPN_1/gadm36_JPN_1.shp')   #To read the shapefile
japanshape
```

```python
japanshape = japanshape[['NL_NAME_1', 'geometry']]    #To keep relevant columnsb
japanshape['pref_name'] = japanshape['NL_NAME_1']
japanshape.plot(figsize=(10, 10))     #To plot the shapefile
```

```python
japanshape = japanshape.set_index('NL_NAME_1')
japanshape
```

```python
import pandas as pd
pref_population = pd.read_csv('./data/pref_population_2015.csv')
STD_PREFS = pref_population['ken'].values
STD_PREFS
```

```python
japanshape = japanshape.reindex(index = STD_PREFS)
japanshape
```

```python
japanshape = japanshape.reset_index()
japanshape
```

```python
japanshape['adjacent_prefs'] = japanshape.apply(lambda x: STD_PREFS[japanshape.geometry.touches(x['geometry'])], axis=1)
japanshape
```

```python
import numpy as np
adjacency_matrix = np.zeros((47, 47), dtype='int32')
for i in range(len(japanshape)):
    adj_pref_index = japanshape.geometry.touches(japanshape.geometry.iloc[i])
    adjacency_matrix[i] = adj_pref_index
np.save('./data/adjacency_matrix.npy', adjacency_matrix)
np.savetxt('./data/adjacency_matrix.csv', adjacency_matrix, fmt='%d')
```
